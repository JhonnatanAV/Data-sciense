```python
# prompt:  NÃºmero de clientes de un gÃ©nero especÃ­fico / NÃºmero total de clientes x 100

# Calculate the percentage of customers of a specific gender
def calculate_percentage_by_gender(df, gender):
  """
  Calculates the percentage of customers of a specific gender in the DataFrame.

  Args:
      df: The DataFrame containing the data.
      gender: The gender to calculate the percentage for.

  Returns:
      The percentage of customers of the specified gender.
  """

  # Count the total number of customers
  total_customers = df.shape[0]

  # Count the number of customers of the specified gender
  gender_count = df[df['Gender'] == gender].shape[0]

  # Calculate the percentage
  percentage = (gender_count / total_customers) * 100

  return percentage

# Calculate the percentage of female customers
percentage_female = calculate_percentage_by_gender(df, 'Female')

# Calculate the percentage of male customers
percentage_male = calculate_percentage_by_gender(df, 'Male')

# Print the results
print(f"Percentage of female customers: {percentage_female:.2f}%")
print(f"Percentage of male customers: {percentage_male:.2f}%")
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Cargar los datos
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PROYECTO/DATA SET/Mall_Customers.csv")

# Seleccionar las caracterÃ­sticas para clustering
features = ['Annual Income (k$)', 'Spending Score (1-100)']

# Crear un modelo KMeans con el nÃºmero de clÃºsteres deseados
n_clusters = 4  # Ajustar segÃºn lo determinado en el anÃ¡lisis
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Entrenar el modelo
kmeans.fit(df[features])

# Asignar etiquetas de clÃºster a los datos
df['Cluster'] = kmeans.labels_

# SegmentaciÃ³n de clientes basada en los clÃºsteres
segmentation = {}
for cluster_id in range(n_clusters):
    cluster_data = df[df['Cluster'] == cluster_id]
    segmentation[f'Cluster {cluster_id}'] = {
        'Average Annual Income': cluster_data['Annual Income (k$)'].mean(),
        'Average Spending Score': cluster_data['Spending Score (1-100)'].mean(),
        'Total Customers': cluster_data.shape[0]
    }

# Imprimir la segmentaciÃ³n de clientes
for cluster, data in segmentation.items():
    print(f"\n{cluster}:")
    print(f"Average Annual Income: {data['Average Annual Income']}")
    print(f"Average Spending Score: {data['Average Spending Score']}")
    print(f"Total Customers: {data['Total Customers']}")

# TambiÃ©n podrÃ­as guardar esta informaciÃ³n en un nuevo archivo CSV
# df.to_csv('segmented_customers.csv', index=False)
```
